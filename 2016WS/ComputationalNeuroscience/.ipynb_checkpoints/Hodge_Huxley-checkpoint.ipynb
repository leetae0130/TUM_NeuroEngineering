# Hodgkin-Huxley Model

Implementation of Hodgkin-Huxley model with python.

Basically, this model describes the dynamics based on electrical properties of neuron.
This model is consists of four variable; m, h, n, u.
Here, m and h describes the conductance of Sodium channel and n describes that of Potassium channel.
U describes the potential change of neuron. 

Alpha and beta values for each gating variable determines the next state value of that gating variable. 
We can check the tendency of opening and closing of the ion channel at each voltage level.

```python
# I got this part from my classmate Steffen during the practical session.
import numpy as np
import matplotlib.pyplot as plt

u_rest = -65
uu = np.linspace(-80, 60, 200)

a_m = lambda u : (2.5 - 0.1*(u - u_rest)) / (np.exp(2.5 - 0.1*(u - u_rest)) - 1)
b_m = lambda u : 4*np.exp(-(u - u_rest)/20)
a_h = lambda u : 0.07*np.exp(-(u - u_rest)/20)
b_h = lambda u : 1/(np.exp(3 - 0.1*(u - u_rest))+1)
g_Na = 120
E_Na = 115 + u_rest

a_n = lambda u : (0.1 - 0.01*(u-u_rest))/(np.exp(1-0.1*(u-u_rest))-1)
b_n = lambda u : 0.125 * np.exp(-(u - u_rest)/80)
g_K = 36
E_K = -12 + u_rest

g_L = 0.3
E_L = 10.6 + u_rest
C = 1

plt.figure(1)


# Plotting the graph for a_m and b_m
plt.subplot(211)
plt.plot(uu, a_m(uu), label="alpha_m")
plt.plot(uu, b_m(uu), label="beta_m")
plt.title("Gating Variable, m")
plt.legend()
plt.show()

# Plotting the graph for a_h and b_h
plt.subplot(211)
plt.plot(uu, a_h(uu), label="alpha_h")
plt.plot(uu, b_h(uu), label="beta_h")
plt.title("Gating Variable, h")
plt.legend()
plt.show()

# Plotting the graph for a_n and b_n
plt.subplot(211)
plt.plot(uu, a_n(uu), label='alpha_n')
plt.plot(uu, b_n(uu), label='beta_n')
plt.title("Gating Variable, n")
plt.legend()
plt.show()
```

# Conductance depends on gating variables

The conductance of potassium channel depends on the value $n^4$.
The sodium channel conductance depends on $n^3*h$.

We plot each value and can check that potassium channel is likely to have high conductance at higher potential.
(You'll learn that it leads to recovery of membrane potential to rest state.)

```python
ddt = lambda alpha, beta, u, last : alpha(u) * (1 - last) - beta(u) * last
update = lambda alpha, beta, u, last, dt : ddt(alpha,  beta, u, last) * dt + last

u_space = np.linspace(-80, 0, 200)
rest_variable = lambda alpha, beta, u_rest : alpha(u_rest)/(alpha(u_rest) + beta(u_rest))

plt.figure(2)

# Plotting the graph for gating variable n
plt.subplot(111)
plt.plot(u_space, rest_variable(a_n, b_n, u_space), label='n')
plt.plot(u_space, rest_variable(a_n, b_n, u_space)**4, label='n**4')
plt.title("Steady State Value, n")
plt.legend()
plt.show()

plt.subplot(111)
plt.plot(u_space, rest_variable(a_m, b_m, u_space), label='m')
plt.plot(u_space, rest_variable(a_h, b_h, u_space), label='h')
plt.plot(u_space, 100*(rest_variable(a_m, b_m, u_space)**3)*rest_variable(a_h, b_h, u_space), label='100*m^3*h')
plt.title("Steady State Value, m and h")
plt.legend()
plt.show()
```

# Experiments on various input conditions

I conducted various experiments using input current generation codes.
You can check out the result in the below.

```python
def simulate(I_in, dt):
    size, = I_in.shape
    
    I_sum = np.zeros(I_in.shape, dtype='float')
    u_t = np.zeros(I_in.shape, dtype='float')
    n_array = np.zeros(I_in.shape, dtype='float')
    m_array = np.zeros(I_in.shape, dtype='float')
    h_array = np.zeros(I_in.shape, dtype='float')
    
    # Initialize the variable, assuming it is in steady state
    n = rest_variable(a_n, b_n, u_rest)
    m = rest_variable(a_m, b_m, u_rest)
    h = rest_variable(a_h, b_h, u_rest)
    u = u_rest
    
    u_t[0] = u_rest
    n_array[0] = n
    m_array[0] = m
    h_array[0] = h
    
    for i in range(1, size):
        current = g_Na*(m**3)*h*(u - E_Na) + g_K*(n**4)*(u - E_K) + g_L*(u - E_L)
        
        du = (I_in[i-1]-current)/C * dt
        u = du + u_t[i-1]
        
        n = update(a_n, b_n, u, n, dt)
        m = update(a_m, b_m, u, m, dt)
        h = update(a_h, b_h, u, h, dt)
        
        u_t[i], I_sum[i], n_array[i], m_array[i], h_array[i] = u, current, n, m, h
    
    return u_t, I_sum, n_array, m_array, h_array
```

```python
def repetitive_input(time, time_rate, initial, value, duration, interval):
    samplesize = int(time * (1/time_rate))
    
    time_space = np.linspace(0, time, samplesize)
    I_output = np.zeros(samplesize)
    I_output[0:samplesize] = initial
    index = 100
    
    while index < time:
        I_output[int((index-duration)*(1/time_rate)):int(index*(1/time_rate))] = initial + value
        index = index + duration + interval
    
    return time_space, I_output

def single_input(time, time_rate, duration, step):
    samplesize = int(time * (1/time_rate))
    
    time_space = np.linspace(0, time, samplesize)
    I_output = np.zeros(samplesize)
    start = int(samplesize*1/2.0)
    
    I_output[start:start+duration*int(1/time_rate)] = step
    
    return time_space, I_output

def step_input(time, time_rate, step, basis=0):
    samplesize = int(time * (1/time_rate))
    
    time_space = np.linspace(0, time, samplesize)
    I_output = np.zeros(samplesize)
    I_output[0:samplesize] = basis
    
    start = int(samplesize*1/2.0)
    
    I_output[start:samplesize-1] = step
    
    return time_space, I_output

def counting_spike(time, time_rate, potential):
    samplesize = int(time * (1/time_rate))
    output_count = np.zeros(samplesize)
    
    for i in range(2, samplesize):
        if(potential[i] > 10 and potential[i-1] < potential[i] and potential[i] > potential[i+1]):
            output_count[i] = 1
    
    return sum(output_count)
```

```python
time = 100
time_rate = 0.01
value = 10

time_r, I_r = single_input(time, time_rate, 4, 5)
u_r, I_rsum, n_r, m_r, h_r = simulate(I_r, time_rate)

fig, ax1 = plt.subplots()
ax1.plot(time_r, I_r, label='I')
ax1.set_ylabel('input current', color = 'b')
ax1.set_ylim([-10,20])
ax2 = ax1.twinx()
ax2.plot(time_r, u_r, label='u', color = 'r')
ax2.set_ylabel('potential')
ax2.set_ylim([-120, 60])
plt.title("Input Current")
plt.legend()
plt.show()
```

```python
# Code for testing refractory period when step size is 3, initial value 3
time = 500
time_rate = 0.01
value = 10

time_r, I_r = repetitive_input(time, time_rate, 0, 3, 30, 50)
u_r, I_rsum, n_r, m_r, h_r = simulate(I_r, time_rate)

fig, ax1 = plt.subplots()
ax1.plot(time_r, I_r, label='I')
ax1.set_ylabel('input current', color = 'b')
ax1.set_ylim([-10,20])
ax2 = ax1.twinx()
ax2.plot(time_r, u_r, label='u', color = 'r')
ax2.set_ylabel('potential')
ax2.set_ylim([-120, 60])
plt.title("Input Current")
plt.legend()
plt.show()

print counting_spike(time, time_rate, u_r)
```

```python
time = 400
time_rate = 0.01
value = 10


for i in [2*x+1 for x in range(5)]:
    time_r, I_r = repetitive_input(time, time_rate, 0, i, 10, 50)
    u_r, I_rsum, n_r, m_r, h_r = simulate(I_r, time_rate)

    fig, ax1 = plt.subplots()
    ax1.plot(time_r, I_r, label='I')
    ax1.set_ylabel('input current', color = 'b')
    ax1.set_ylim([-10,20])
    ax2 = ax1.twinx()
    ax2.plot(time_r, u_r, label='u', color = 'r')
    ax2.set_ylabel('potential')
    ax2.set_ylim([-120, 60])
    plt.title("Input Current")
    plt.legend()
    plt.show()
    print "The number of spikes : " + str(counting_spike(time, time_rate, u_r))
```

```python
time = 400
time_rate = 0.01
value = 10


for i in [2*x+1 for x in range(5)]:
    time_r, I_r = repetitive_input(time, time_rate, 0, 5, 15, i*15)
    u_r, I_rsum, n_r, m_r, h_r = simulate(I_r, time_rate)

    plt.figure(1)
    plt.subplot(111)
    plt.plot(time_r, I_r, label='I')
    plt.plot(time_r, u_r, label='u')
    plt.title("Action Potential Period of " + str(i))
    plt.legend()
    plt.show()
    
    print "The number of spikes : " + str(counting_spike(time, time_rate, u_r))

```

# Gradually increasing current doesn't lead to spike.

```python
# Gradually Increasing Current
time = 100
time_rate = 0.05
start = 1000

samplesize = int(time * (1/time_rate))
grad = lambda u, b0, b1 : u*b1 + b0
    
time_space = np.linspace(0, time, samplesize)
I_output = grad(time_space, 0.1, 0.02)
I_output[start:samplesize-1] = 0

u_r, I_rsum, n_r, m_r, h_r = simulate(I_output, time_rate)

fig, ax1 = plt.subplots()
ax1.plot(time_space, I_output, label='I')
ax1.set_ylabel('input current', color = 'b')
ax1.set_ylim([-5, 20])
ax2 = ax1.twinx()
ax2.plot(time_space, u_r, label='u', color = 'r')
ax2.set_ylabel('potential')
ax2.set_ylim([-120, 60])
plt.title("Input Current")
plt.legend()
plt.show()
```
