# Basic Simulator for Entropy

In the code below, I will plot the value of binary entropy for given probability.  
We can simply observe that the value of entropy gets to the maximum value 1 when the probability is 1/2.

```python
import math
import matplotlib.pyplot as plt
from matplotlib import cm
import mpl_toolkits.mplot3d.axes3d as axes3d
import numpy as np

p_logp = lambda p : p*math.log(1/p, 2) if p > 0 and p < 1 else 0
b_entropy = lambda a : p_logp(a) + p_logp(1-a)
entropy = lambda A : sum([p_logp(a) for a in A])
combination = lambda n, p : math.factorial(n)/(math.factorial(int(n*p))*math.factorial(int(n*(1-p))))

x_b = np.linspace(0, 1, 200)
y_b = [b_entropy(x) for x in x_b]

plt.figure(1)
plt.subplot(111)
plt.xlabel('probability')
plt.ylabel('entropy')
plt.ylim([0, 1.3])
plt.plot(x_b, y_b, label='binary_entropy')
plt.show()
```

As entropy is driven from the combination value $$\log\binom{n}{np} = \log\frac{n!}{np!n(1-p)!}$$,  
We can check that whether $\log\binom{n}{np}$ converges to $\mathrm{H(p)}$ as n becomes larger and larger.  

As we can see in the below plot, the difference between those two numbers converges to 0 as n becomes bigger.

```python
p = 1/2.0
half = 1/2.0
third = 1/3.0
hundred = 1/100.0
x_half = np.linspace(1, 1000, 500)
y_half = [math.log(combination(int(x), half), 2)/x for x in x_half]
y_third = [math.log(combination(int(x), third), 2)/x for x in x_half]
y_hundred = [math.log(combination(int(x), hundred), 2)/x for x in x_half]
xx = np.linspace(1, 1000, 500)

plt.figure(1)
plt.title('log of combination')
plt.subplot(111)
plt.plot(x_half, y_half, label="log(n, n*1/2) base of 2")
plt.plot(x_half, y_third, label="log(n, n*1/3) base of 2")
plt.plot(x_half, y_hundred, label="log(n, n*1/10) base of 2")
#plt.plot(x_half, xx, label="x=x")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

entropy_two = b_entropy(1/2.0)
entropy_three = b_entropy(1/3.0)
entropy_hundred = b_entropy(1/100.0)
diff_two = [entropy_two - y for y in y_half]
diff_three = [entropy_three - y for y in y_third]
diff_hundred = [entropy_hundred - y for y in y_hundred]

plt.figure(2)
plt.title('diffenrence between log of combination and entropy')
plt.subplot(111)
plt.plot(x_half, diff_two, label="log(n, n*1/2) base of 2")
plt.plot(x_half, diff_three, label="log(n, n*1/3) base of 2")
plt.plot(x_half, diff_hundred, label="log(n, n*1/10) base of 2")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
